```python
prop = property.drop(['PID', 'CM_ID', 'GIS_ID', 'ST_NUM', 'ST_NAME', 'UNIT_NUM', 'BLDG_SEQ', 'NUM_BLDGS', 'LUC', 'LU_DESC', 'OWN_OCC', 'OWNER', 'MAIL_ADDRESSEE', 'MAIL_ADDRESS', 'MAIL_CITY', 'MAIL_ZIPCODE', 'RES_UNITS', 'COM_UNITS', 'RC_UNITS', 'CD_FLOOR', 'GROSS_AREA', 'BDRM_COND', 'OVERALL_COND', 'INT_WALL', 'EXT_FINISHED', 'INT_COND', 'EXT_COND', 'BTHRM_STYLE1', 'BTHRM_STYLE2', 'KITCHEN_TYPE', 'BTHRM_STYLE3', 'KITCHEN_STYLE1', 'KITCHEN_STYLE2', 'KITCHEN_STYLE3', 'HEAT_FUEL', 'ORIENTATION', 'PROP_VIEW', 'CORNER_UNIT'], axis=1, inplace = False)
```

```python
# Small multiple of building value by construction year in multiple locations and house types.
```

```python
prop1 = prop.dropna()
prop2 = prop1.groupby(['CITY']).agg({col: ['size'] for col in prop1.columns})
prop2.columns = ['_'.join(multi_index) for multi_index in prop2.columns.ravel()]
prop2 = prop2.reset_index()
prop2.head()
```

```python
prop3 = prop1.loc[~(prop1['CITY'].isin(['DEDHAM', 'CHESTNUT HILL', 'BROOKLINE']))]
prop3.head()
```

```python
prop4 = prop3.groupby(['CITY']).agg({col: ['size'] for col in prop3.columns})
prop4.columns = ['_'.join(multi_index) for multi_index in prop4.columns.ravel()]
prop4 = prop4.reset_index()
prop4['CITY_size'] = pd.to_numeric(prop4['CITY_size'], downcast='integer', errors='coerce')
prop4 = prop4.sort_values(by=['CITY_size'], ascending=[False])
fig = px.bar(prop4, x='CITY_size', y='CITY', template='plotly_white', title='Sample sizes of city data')
fig.update_xaxes(title_text='Count of properties')
fig.update_yaxes(title_text='yAxis title')
fig
```

```python
fig = px.histogram(prop3, x='LU', template='xgridoff', title='Histogram of Land Use type')
fig
```

```python
prop4 = prop3.loc[prop3['LU'].isin(['R3', 'R1', 'R2'])]
```

```python
prop4['LAND_VALUE'] = prop4['LAND_VALUE'].replace(',','')
# Step: Manipulate strings of 'LAND_VALUE' via Find ',' and Replace with ''
prop4["LAND_VALUE"] = prop4["LAND_VALUE"].str.replace(',', '', regex=False)

# Step: Manipulate strings of 'BLDG_VALUE' via Find ',' and Replace with ''
prop4["BLDG_VALUE"] = prop4["BLDG_VALUE"].str.replace(',', '', regex=False)

# Step: Manipulate strings of 'TOTAL_VALUE' via Find ',' and Replace with ''
prop4["TOTAL_VALUE"] = prop4["TOTAL_VALUE"].str.replace(',', '', regex=False)

# Step: Manipulate strings of 'GROSS_TAX' via Find ',' and Replace with ''
prop4["GROSS_TAX"] = prop4["GROSS_TAX"].str.replace(',', '', regex=False)

# Step: Manipulate strings of 'GROSS_TAX' via Find '$' and Replace with ''
prop4["GROSS_TAX"] = prop4["GROSS_TAX"].str.replace('$', '', regex=False)

# Step: Change data type of LAND_VALUE to Integer
prop4['LAND_VALUE'] = pd.to_numeric(prop4['LAND_VALUE'], downcast='integer', errors='coerce')

# Step: Change data type of BLDG_VALUE to Integer
prop4['BLDG_VALUE'] = pd.to_numeric(prop4['BLDG_VALUE'], downcast='integer', errors='coerce')

# Step: Change data type of TOTAL_VALUE to Integer
prop4['TOTAL_VALUE'] = pd.to_numeric(prop4['TOTAL_VALUE'], downcast='integer', errors='coerce')

# Step: Change data type of GROSS_TAX to Float
prop4['GROSS_TAX'] = pd.to_numeric(prop4['GROSS_TAX'], downcast='float', errors='coerce')

# Step: Change data type of YR_BUILT to Integer
prop4['YR_BUILT'] = pd.to_numeric(prop4['YR_BUILT'], downcast='integer', errors='coerce')

# Step: Change data type of YR_REMODEL to Integer
prop4['YR_REMODEL'] = pd.to_numeric(prop4['YR_REMODEL'], downcast='integer', errors='coerce')
```

```python
prop5 = prop4
fig = px.histogram(prop5.sample(n=10000, replace=False, random_state=123).sort_index(), x='LAND_VALUE', range_x=(0, 1000000), template='xgridoff', title='Histogram of Land Values')
fig.update_yaxes(title_text='Land Value')
fig
```

```python
fig = px.histogram(prop5.sample(n=10000, replace=False, random_state=123).sort_index(), facet_row='LU', x='TOTAL_VALUE', range_x=(0, 2000000), range_y=(0, 500), title='Histogram of Total Values of 1, 2, and 3 bed properties')
fig
```

```python
prop5["COST_PerAREA"] = prop5['LAND_VALUE']/prop5['LAND_SF']
```

```python
prop6 = prop5
import pandas as pd; import numpy as np
# Step: Group by and aggregate
prop6 = prop5.groupby(['CITY']).agg(COST_PerAREA_max=('COST_PerAREA', 'max')).reset_index()

# Step: Sort column(s) COST_PerAREA_max descending (Z-A)
prop6 = prop6.sort_values(by=['COST_PerAREA_max'], ascending=[False])

import plotly.express as px
fig = px.line(prop6, x='CITY', y='COST_PerAREA_max', title='Maximum land cost per unit area in different cities', labels={'BOSTON': 'BOS', 'CHARLESTOWN': 'CHAR', 'SOUTH BOSTON': 'S.BOS', 'ROXBURY CROSSING': 'ROX.C', 'EAST BOSTON': 'E.BOS', 'ALLSTON': 'ALLS', 'JAMAICA PLAIN': 'JP', 'DORCHESTER': 'DOR', 'BRIGHTON': 'BRIG', 'ROXBURY': 'ROX', 'ROSLINDALE': 'RSLN', 'WEST ROXBURY': 'W.ROX', 'MATTAPAN': 'MATT', 'HYDE PARK': 'HP'})
fig.update_yaxes(title_text='Cost per unit area')
fig
```

```python
# Step: Group by and aggregate
prop6 = prop5.groupby(['CITY']).agg({'COST_PerAREA': ['mean', 'min', 'max']})
prop6.columns = ['_'.join(multi_index) for multi_index in prop6.columns.ravel()]
prop6 = prop6.reset_index()

# Step: Sort column(s) COST_PerAREA_max descending (Z-A)
prop6 = prop6.sort_values(by=['COST_PerAREA_max'], ascending=[False])
city_dict ={'BOSTON': 'BOS', 'CHARLESTOWN': 'CTWN', 'SOUTH BOSTON': 'S.BOS', 'ROXBURY CROSSIN': 'ROX.C', 'EAST BOSTON': 'E.BOS', 'ALLSTON': 'ALLS', 'JAMAICA PLAIN': 'JP', 'DORCHESTER': 'DOR', 'BRIGHTON': 'BRIG', 'ROXBURY': 'ROX', 'ROSLINDALE': 'ROS', 'WEST ROXBURY': 'W.ROX', 'MATTAPAN': 'MTPN', 'HYDE PARK': 'HP'}
prop6 = prop6.replace({"CITY": city_dict})
prop6
```

```python
fig = px.line(prop6, x='CITY', y=['COST_PerAREA_mean', 'COST_PerAREA_max'], template='simple_white', title='Mean and Max cost per unit area by city')
fig.update_xaxes(tickangle=0)
fig.update_yaxes(title_text='Cost per unit area')
fig
```

```python
fig = px.histogram(prop5, x='LIVING_AREA', facet_col = "CITY"
             ,facet_row = 'LU', title='Distribution of Living Area for 1, 2, and 3 family houses in different cities', marginal = 'rug', labels={
                     "LIVING_AREA": "Area"
                 }, template='simple_white', height=600, width=1800, opacity=0.7, range_x=[0, 7000], range_y=[0, 250], nbins=200)
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.show()
```
